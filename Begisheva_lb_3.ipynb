## 1 Задача 1.3.12
Знайти ймовiрнiсть того, що серед трьох цифр, кожна з яких вибрана навмання, буде лише 1, 2, 3 рiзних.

Обчисліть теоретичну (повним перебором) та еміричну (симулюванням $100000$ експериментів) імовірності.

```python
# Місце для Вашого коду
from itertools import *
import numpy as np

def teor(n,dif):
    p=0
    p_all = 0
    for x in product(range(10), repeat=n):
        if len(set(x))==dif:
            p+=1
        p_all+=1
    return p/p_all

def emper(N,n,dif):
    p=0
    for _ in range(N):
        x = np.random.randint(10, size=n)
        if len(set(x))==dif:
            p+=1
    return p/N
n=3
k=[1,2,3]
N=100000
for i in k:
    print(f"teor({n},{i})=",teor(n,i))
    print(f"emper({n},{i})=",emper(N,n,i))

```

# 2 Задача 1.3.14
З послiдовностi чисел $1, 2, . . . , n$ вибирають навмання $k$ рiзних чисел.
Яка ймовiрнiсть того, що:
1. кожне з вибраних чисел кратне даному числу $p$;
2. кожне з вибраних чисел кратне хоча б одному з двох взаємно простих чисел $p$ i $q$;
3. серед вибраних чисел є хоча б одне кратне $p$?

Напишіть  відповідні функції для обрахунку теоретичної (повним перебором) та еміричної (симулюванням $100000$ експериментів) імовірностей в залежності від параметрів $n, k, p, q$.
Виведіть результат для
- $n = 25, k = 5, p = 3, q = 4$;
- $n = 25, k = 10, p = 3, q = 4$.

```python
# Місце для Вашого коду
from itertools import *
import numpy as np
def fun(x,p,q):
    ok=[1,1,0]
    for i in x:
        if i%p!=0:
            ok[0]=0
        if i%p!=0 and i%q!=0:
            ok[1]=0
        if i%p==0:
            ok[2]=1
        if ok==[0,0,1]:
            break
    return ok
    
def teor(n,k,p,q):
    t=[0,0,0]
    p_all = 0
    for x in combinations(range(1,n+1),k):
        ok=fun(x,p,q)
        for i in range(len(t)):
            t[i]+=ok[i]
        p_all+=1
    return [t[0]/p_all,t[1]/p_all,t[2]/p_all]

def emper(N,n,k,p,q):
    t=[0,0,0]
    for _ in range(N):
        x = np.random.choice(range(1, n+1), size=k, replace=False)
        ok=fun(x,p,q)
        for i in range(len(t)):
            t[i]+=ok[i]
    return [t[0]/N,t[1]/N,t[2]/N]
n=25
k=[5,10]
p=3
q=4
N=100000
for i in k:
    print(f"teor({n},{i}) =",teor(n,i,p,q))
    print(f"emper({n},{i}) =",emper(N,n,i,p,q))
```

# 3 Задача 1.4.6
В коморi знаходяться $n$ пар черевикiв.
З них випадковим чином вибираються $2k$ черевикiв.
Яка ймовiрнiсть того, що серед вибраних черевикiв:
1. вiдсутнi парнi;
2. є рiвно одна комплектна пара;
3. є двi комплектнi пари?

Напишіть  відповідні функції для обрахунку теоретичної (повним перебором) та еміричної (симулюванням $100000$ експериментів) імовірностей в залежності від параметрів $n, k$.
Виведіть результат для
- $n = 8, k = 4$;
- $n = 13, k = 5$.

```python
# Місце для Вашого коду
from itertools import *
import numpy as np

def teor(n,k):
    t=[0,0,0]
    p_all = 0
    for x in combinations(list(range(n))+list(range(n)), 2*k):
        if len(set(x))==len(x):
            t[0]+=1
        elif len(set(x))==len(x)-1:
            t[1]+=1
        elif len(set(x))<=len(x)-2:
            t[2]+=1
        p_all+=1
    return [t[0]/p_all,t[1]/p_all,t[2]/p_all]

def emper(N,n,k):
    t=[0,0,0]
    for _ in range(N):
        x = np.random.choice(list(range(n))+list(range(n)), size=2*k, replace=False)
        if len(set(x))==len(x):
            t[0]+=1
        elif len(set(x))==len(x)-1:
            t[1]+=1
        elif len(set(x))<=len(x)-2:
            t[2]+=1
    return [t[0]/N,t[1]/N,t[2]/N]
n=[8,13]
k=[4,5]
N=100000
for i in range(len(k)):
    print(f"teor({n[i]},{k[i]})=",teor(n[i],k[i]))
    print(f"emper({n[i]},{k[i]})=",emper(N,n[i],k[i]))
```

Знайти цi ймовiрностi.
Знайти ймовiрнiсть того, що в результатi експерименту з’явиться:
1. парне число;
2. непарне число.

Напишіть функцію для обрахунку еміричної (симулюванням $100000$ експериментів) імовірності в залежності від параметра $n$.
Виведіть результат для
- $n = 10$;
- $n = 25$.

```python
# Місце для Вашого коду
from itertools import *
import numpy as np
from math import log

def teor(N,n):
    arr=[]
    probs=[]
    p1=0
    p2=0
    p=sum(log(i) for i in range(1,2*n+1))
    for i in range(1,2*n+1):
        p0=log(i)/p
        arr.append(i)
        probs.append(p0)
        if i%2==0:
            p1+=p0
        else:
            p2+=p0
    print(n,"--- even:",p1," odd:",p2)
  
    p1=0
    p2=0
    for i in range(N):
        rand=np.random.choice(arr, size=1, replace=True, p=probs)
        if rand%2==0:
            p1+=1
        else:
            p2+=1
    print(n,"empir --- even:",p1/N," odd:",p2/N)

N=100000
n=[10,25]
for i in n:
    teor(N,i)
```
